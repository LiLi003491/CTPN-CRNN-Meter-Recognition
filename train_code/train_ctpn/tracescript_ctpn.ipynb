```python
import torch
from ctpn_model import CTPN_Model
import os

model_path = './ctpn_models'
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
weights = os.path.join(model_path, 'CTPN.pth')

model = CTPN_Model()
model.load_state_dict(torch.load(weights, map_location=device)['model_state_dict'])
model.to(device)
model.eval()
print(model)
example = torch.rand(1, 3, 416, 512)
traced_script_module = torch.jit.trace(model, example)
traced_script_module.save("./ctpn_models/ctpn_script.pt")
```
