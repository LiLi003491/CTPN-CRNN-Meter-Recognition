```python
import torch
import keys
import crnn as crnn

alphabet = keys.alphabet
model_path = './crnn_models/CRNN-7149_302_989.pth'

model = crnn.CRNN(32, 1, len(alphabet) + 1, 256) 

model.load_state_dict(torch.load( model_path, map_location=torch.device('cpu')))
# for param_tensor in model.state_dict():
#     print(param_tensor,'\t',model.state_dict()[param_tensor].size())
#     print(type(param_tensor))
#     print(type(model.state_dict()[param_tensor]))
#     print(type(model.state_dict()))
model.eval()
example = torch.rand(1, 1, 32, 512)
traced_script_module = torch.jit.trace(model, example)
print(type(traced_script_module))
print(traced_script_module.code)
traced_script_module.save("./crnn_models/decimal_script.pt")
```

```python
from crnn_recognizer import resizeNormalize, strLabelConverter, PytorchOcr
import os
import cv2
model_path = './crnn_models/CRNN-7149_302_989.pth'
recognizer = PytorchOcr(model_path)

def read_template(directory_name):

#         imgList = os.listdir(r"./"+directory_name)
    imgList = os.listdir(directory_name)
    imgList.sort(key=lambda x: int(x.split('.')[0])) 

    for count in range(0, len(imgList)):
        filename = imgList[count]
        img = cv2.imread(directory_name + "/" + filename)
        res = recognizer.recognize(img)
        print(res)

read_template('E:/Jupyter/yolov5_crnn/runs/detect/exp2/crops/number')
```
